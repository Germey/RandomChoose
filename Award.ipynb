```python
import math
import random
import pandas as pd
import json
```

```python
def choose(keys, weights, number):
    result = []
    weights = list(map(lambda x: float(x), weights))
    for i in range(len(keys)):
        k = math.pow(random.random(), 1 / weights[i])
        if len(result) < number:
            result.append((keys[i], k))
        else:
            min_element = min(result, key=lambda x: x[1])
            if k > min_element[1]:
                result.remove(min_element)
                result.append((keys[i], k))
    return list(map(lambda x: x[0], result))
```

```python
df = pd.read_csv('data.csv', dtype=object)
df
```

```python
names, sns = df['name'].tolist(), df['sn'].tolist()
keys = list(zip(names, sns))
print('Keys Sample:', keys[0:5])
print('Number of Keys:', len(keys))
```

```python
weights = df['lucky'].tolist()
print('Weights Sample:', weights[0:5])
print('Number of Keys:', len(weights))
```

```python
number = 30
```

```python
result = choose(keys, weights, number)
print(result)
print('Number of Results:', len(result))
```
